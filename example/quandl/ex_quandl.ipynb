## データセットのformatデータ取得

各データにはデータごとに固有情報を持っているので、それを確認する必要がある
* 頻度(日次・週次・年次)
* 最終更新日
* データ系列名・項目名(close, openなど)

```python
import pandas as pd
import io
import requests
```

```python
bef_url = "https://www.quandl.com/api/v3/datasets.csv?database_code=CHRIS&per_page=100&sort_by=id&page="
aft_url = "&api_key=CbzNMfBFSFEFsiNNAe97"
page_num = 40
url = [bef_url + str(x) + aft_url for x in range(page_num)] # フォーマットファイルのダウンロードURL
```

```python
# 内包表記使えるように関数化
def call_format(url):
    s=requests.get(url).content
    return pd.read_csv(io.StringIO(s.decode('utf-8')))
```

```python
df_chris = [call_format(url[x]) for x in range(page_num)] # フォーマットファイルを1つずつダウンロードしてリストに含める
```

```python
df_chris_all = pd.concat(df_chris, axis=0) # リスト内のdataframeを縦に繋ぐ
df_chris_all = df_chris_all.reset_index() # index番号振り直し
df_chris_all
```

```python
df_chris_all.to_csv("chris_all.csv") # CSV出力
```

## フォーマットデータから時系列データ取得 

```python
# frequency 
df_chris_daily = df_chris_all.query("frequency == 'daily'") # frequencyがdailyのものだけフィルタ
df_chris_daily = df_chris_daily.reset_index()
```

```python
df_chris_daily.tail(5)
```

```python
# 利用可能データのYYYY-MM-DDのうちYYYYを取得し、新しい列に追加
df_chris_daily = df_chris_daily.assign(
    new_month = [df_chris_daily.newest_available_date[x].split("-")[0] for x in range(df_chris_daily.shape[0])])
```

```python
df_chris_daily
```

```python
df_chris_daily_2016 = df_chris_daily.query("new_month == '2016'") # 最新データが2016年のものだけ
```

```python
# ダウンロード用ID(データセットコード/データベースコード)作成
temp_pd = df_chris_daily_2016[['dataset_code','database_code']]
pd_ch = temp_pd.assign(ql_tick =  temp_pd['database_code'] + '/' + temp_pd['dataset_code'])
pd_ch = pd_ch.reset_index()
pd_ch
```

```python
# 暫定的に50個データ取得
# 一度にget出来るデータ系列数は限られているようなので、ループさせるようにしないとダメ
# 100系列同時だとダメだった->50系列なら行けた
l_ch = [ql.get(pd_ch.ql_tick[x] + ".4", start_date='2001-01-01') for x in range(50)]# column_nameの何番目かを指定すれば、そのデータ列だけ落とせる
pd_ch_f = pd.concat(l_ch, axis=1)
```
