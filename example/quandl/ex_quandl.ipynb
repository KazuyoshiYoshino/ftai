## とりあえずデータ取得の方法

getでデータしょっぴける.

基本的な使い方は公式参照：https://www.quandl.com/

```python
df_GDP = ql.get('ODA/JPN_NGDP_R') # 日本GDP
df_GDP.tail(10)
```

```python
df_STOCK = ql.get(['NIKKEI/INDEX', 'YAHOO/INDEX_GSPC']) # 日経平均株価、S&P株価
df_STOCK.tail(15)
```
